```python
# ===============================================================
# 📘 Project: Crop Yield Estimation using Big Data Analytics
# Demonstrating Applications of Big Data in the Agriculture Industry
# ===============================================================

# Step 1ï¸âƒ£ : Import Required Libraries
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, round, monotonically_increasing_id
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
import matplotlib.pyplot as plt

# Initialize Spark Session
spark = SparkSession.builder.appName("Crop_Yield_Estimation_BigData").getOrCreate()

# Step 2ï¸âƒ£ : Load Data from Databricks Table
df = spark.table("workspace.default.agriculture")
print("âœ… Dataset Loaded Successfully from workspace.default.agriculture")
display(df)

# Step 3ï¸âƒ£ : Select Features and Target Column
feature_cols = [
    "Rainfall_mm",
    "Fertilizer_kg_per_acre",
    "Soil_Health_Index",
    "Temperature_C",
    "Humidity_%",
    "Soil_Moisture_%",
    "Irrigation_Level_%",
    "Pesticide_Litre_per_acre",
    "Seed_Quality_Index"
]
target_col = "Yield_ton_per_acre"

# Prepare DataFrame for modeling
df_model = df.select(feature_cols + [target_col]).na.drop()
print("âœ… Features selected for yield prediction")
display(df_model.limit(10))

# Step 4ï¸âƒ£ : Assemble Features and Split Data
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
assembled_df = assembler.transform(df_model).select("features", target_col)
train_data, test_data = assembled_df.randomSplit([0.8, 0.2], seed=42)
print(f"Training Records: {train_data.count()} | Testing Records: {test_data.count()}")

# Step 5ï¸âƒ£ : Train Linear Regression Model
lr = LinearRegression(featuresCol="features", labelCol=target_col)
lr_model = lr.fit(train_data)
print("âœ… Linear Regression Model Trained Successfully")

# Step 6ï¸âƒ£ : Evaluate Model Performance
predictions = lr_model.transform(test_data)
evaluator = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="r2")
r2 = evaluator.evaluate(predictions)
rmse = lr_model.summary.rootMeanSquaredError
print(f"📊 Model Performance â†’ RÂ² = {r2:.3f}, RMSE = {rmse:.3f}")

# Step 7ï¸âƒ£ : Predict on Full Dataset (Fixed for Duplicate Columns)
full_predictions = lr_model.transform(assembled_df)

# Add unique IDs to both dataframes
df_indexed = df_model.withColumn("id", monotonically_increasing_id())

# Select only 'prediction' from full_predictions to avoid duplicate columns
preds_indexed = full_predictions.withColumn("id", monotonically_increasing_id()).select("id", "prediction")

# Join both safely
final_df = df_indexed.join(preds_indexed, on="id", how="inner")

# Rename and clean column names
final_df = final_df.withColumnRenamed("Yield_ton_per_acre", "Actual_Yield_ton_per_acre") \
                   .withColumnRenamed("prediction", "Predicted_Yield_ton_per_acre")

# âœ… Drop duplicates (ensure every column name is unique)
from pyspark.sql import DataFrame

def drop_duplicate_columns(df: DataFrame):
    cols = df.columns
    seen = set()
    unique_cols = []
    for c in cols:
        if c not in seen:
            unique_cols.append(c)
            seen.add(c)
    return df.select(*[col(c) for c in unique_cols])

final_df = drop_duplicate_columns(final_df)

# Round values
for c in ["Actual_Yield_ton_per_acre", "Predicted_Yield_ton_per_acre"]:
    final_df = final_df.withColumn(c, round(col(c), 2))

print("âœ… Final Predictions Generated (No Duplicate Columns)")
display(final_df.limit(10))

# Step 8ï¸âƒ£ : Save Clean Predictions Table
final_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("workspace.default.crop_yield_predictions")

print("📂 Predictions table saved successfully: workspace.default.crop_yield_predictions")

# Step 9ï¸âƒ£ : Dashboard Analytics
avg_yield_district = df.groupBy("District").avg("Yield_ton_per_acre").orderBy("District")
display(avg_yield_district)

avg_yield_crop = df.groupBy("Crop_Type").avg("Yield_ton_per_acre").orderBy("Crop_Type")
display(avg_yield_crop)

avg_yield_pred = final_df.groupBy().avg("Actual_Yield_ton_per_acre", "Predicted_Yield_ton_per_acre")
display(avg_yield_pred)

# Step 🔟 : Visualization (Actual vs Predicted)
pred_pd = final_df.select("Actual_Yield_ton_per_acre", "Predicted_Yield_ton_per_acre").toPandas()
plt.figure(figsize=(7,7))
plt.scatter(pred_pd["Actual_Yield_ton_per_acre"], pred_pd["Predicted_Yield_ton_per_acre"], color="green", alpha=0.6)
plt.xlabel("Actual Yield (tons/acre)")
plt.ylabel("Predicted Yield (tons/acre)")
plt.title("Actual vs Predicted Crop Yield")
plt.grid(True)
display(plt)

# Step 1ï¸âƒ£1ï¸âƒ£ : KPI Dashboard
kpi_df = spark.createDataFrame([(r2, rmse)], ["R2_Score", "RMSE"])
display(kpi_df)

print("âœ… Model KPIs and Dashboard Outputs Ready")
```

```python
# ===============================================================
# 📘 PROJECT: Agricultural Commodity Price Forecasting using Big Data Analytics
# Demonstrating Applications of Big Data in Agriculture
# Author: Krishna Chaitanya
# ===============================================================

# Step 1ï¸âƒ£ : Import Required Libraries
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, avg, round, monotonically_increasing_id, lag
from pyspark.sql.window import Window
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
import matplotlib.pyplot as plt

# Initialize Spark Session
spark = SparkSession.builder.appName("Commodity_Price_Forecasting_BigData").getOrCreate()

# Step 2ï¸âƒ£ : Load Data from Databricks Table
df = spark.table("workspace.default.agriculture")
print("âœ… Dataset Loaded Successfully")
display(df)

# Step 3ï¸âƒ£ : Select Relevant Columns for Forecasting
# --------------------------------------------------
price_df = df.select(
    "Year",
    "Crop_Type",
    "Rainfall_mm",
    "Temperature_C",
    "Humidity_%",
    "Average_Price_INR"
).na.drop()

display(price_df.limit(10))

# Step 4ï¸âƒ£ : Aggregate Average Prices per Crop per Year
# -----------------------------------------------------
avg_price_df = price_df.groupBy("Crop_Type", "Year").avg(
    "Rainfall_mm", "Temperature_C", "Humidity_%", "Average_Price_INR"
).withColumnRenamed("avg(Average_Price_INR)", "Avg_Price_INR") \
 .withColumnRenamed("avg(Rainfall_mm)", "Avg_Rainfall_mm") \
 .withColumnRenamed("avg(Temperature_C)", "Avg_Temperature_C") \
 .withColumnRenamed("avg(Humidity_%)", "Avg_Humidity_%")

print("âœ… Aggregated yearly crop prices successfully")
display(avg_price_df.orderBy("Crop_Type", "Year"))

# Step 5ï¸âƒ£ : Create Lag Feature to Capture Price Trends
# -----------------------------------------------------
windowSpec = Window.partitionBy("Crop_Type").orderBy("Year")
avg_price_df = avg_price_df.withColumn("Prev_Year_Price", lag("Avg_Price_INR", 1).over(windowSpec))
avg_price_df = avg_price_df.fillna({"Prev_Year_Price": 0})

# Step 6ï¸âƒ£ : Assemble Features for Machine Learning
# -------------------------------------------------
feature_cols = ["Avg_Rainfall_mm", "Avg_Temperature_C", "Avg_Humidity_%", "Prev_Year_Price"]
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
assembled_df = assembler.transform(avg_price_df).select("features", "Avg_Price_INR", "Crop_Type", "Year")

# Step 7ï¸âƒ£ : Split Data into Training and Testing
train_data, test_data = assembled_df.randomSplit([0.8, 0.2], seed=42)

# Step 8ï¸âƒ£ : Train the Linear Regression Model
lr = LinearRegression(featuresCol="features", labelCol="Avg_Price_INR")
lr_model = lr.fit(train_data)
print("âœ… Linear Regression Model Trained Successfully")

# Step 9ï¸âƒ£ : Evaluate Model Performance
from pyspark.ml.evaluation import RegressionEvaluator
predictions = lr_model.transform(test_data)
evaluator = RegressionEvaluator(labelCol="Avg_Price_INR", predictionCol="prediction", metricName="r2")
r2 = evaluator.evaluate(predictions)
rmse = lr_model.summary.rootMeanSquaredError
print(f"📊 Model Performance â†’ RÂ² = {r2:.3f}, RMSE = {rmse:.3f}")

# Step 🔟 : Predict Prices for Full Dataset
# -----------------------------------------
full_predictions = lr_model.transform(assembled_df)

final_df = full_predictions.withColumnRenamed("Avg_Price_INR", "Actual_Price_INR") \
                           .withColumnRenamed("prediction", "Predicted_Price_INR")

final_df = final_df.withColumn("Actual_Price_INR", round(col("Actual_Price_INR"), 2)) \
                   .withColumn("Predicted_Price_INR", round(col("Predicted_Price_INR"), 2))

display(final_df.orderBy("Crop_Type", "Year"))

# Step 1ï¸âƒ£1ï¸âƒ£ : Save Forecasted Results to Delta Table
final_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("workspace.default.commodity_price_forecast")
print("📂 Forecast results saved successfully as: workspace.default.commodity_price_forecast")

# Step 1ï¸âƒ£2ï¸âƒ£ : Generate Analytical Summaries for Dashboard
# ----------------------------------------------------------
# Crop-wise comparison
avg_pred_crop = final_df.groupBy("Crop_Type").avg("Actual_Price_INR", "Predicted_Price_INR").orderBy("Crop_Type")
display(avg_pred_crop)

# Year-wise comparison
avg_pred_year = final_df.groupBy("Year").avg("Actual_Price_INR", "Predicted_Price_INR").orderBy("Year")
display(avg_pred_year)

# Step 1ï¸âƒ£3ï¸âƒ£ : Visualization - Actual vs Predicted Commodity Prices
# -------------------------------------------------------------------
pred_pd = final_df.select("Actual_Price_INR", "Predicted_Price_INR").toPandas()

plt.figure(figsize=(7,7))
plt.scatter(pred_pd["Actual_Price_INR"], pred_pd["Predicted_Price_INR"], color="orange", alpha=0.6)
plt.xlabel("Actual Commodity Price (â‚¹/quintal)")
plt.ylabel("Predicted Commodity Price (â‚¹/quintal)")
plt.title("Actual vs Predicted Commodity Prices")
plt.grid(True)
display(plt)

# Step 1ï¸âƒ£4ï¸âƒ£ : KPI Table for Dashboard
kpi_df = spark.createDataFrame([(r2, rmse)], ["R2_Score", "RMSE"])
display(kpi_df)

print("âœ… Commodity Price Forecasting Completed Successfully and Dashboard Data Ready")
```

```python
# ===============================================================
# 📘 Project: Precision Farming using IoT & Smart Sensors
# Demonstrating Applications of Big Data Analytics in Agriculture
# Compatible with Databricks Serverless & All-purpose Clusters
# ===============================================================

# Step 1ï¸âƒ£ : Import Required Libraries
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, lit, rand, avg, stddev, abs, count, sum, round
from pyspark.sql.types import DoubleType, IntegerType
from pyspark.sql.window import Window
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
import matplotlib.pyplot as plt

# Initialize Spark Session
spark = SparkSession.builder.appName("Precision_Farming_IoT_Sensors").getOrCreate()

# Step 2ï¸âƒ£ : Load Dataset
input_table = "workspace.default.agriculture"
df = spark.table(input_table)
print(f"âœ… Dataset loaded successfully: {input_table}")
display(df.limit(10))

# Step 3ï¸âƒ£ : Ensure Required IoT Sensor Columns Exist (Simulate if Missing)
expected_cols = {
    "Sensor_Temperature_C": ("Temperature_C", lambda: 25 + rand()*10),
    "Sensor_Soil_Moisture_%": ("Soil_Moisture_%", lambda: 30 + rand()*20),
    "NDVI_Index": ("NDVI_Index", lambda: 0.3 + rand()*0.6),
    "PH_Value": ("PH_Value", lambda: 6.0 + rand()*2.0),
    "Irrigation_Level_%": ("Irrigation_Level_%", lambda: 40 + rand()*50)
}

tmp = df
for col_name, (source_col, expr) in expected_cols.items():
    if col_name not in tmp.columns:
        print(f"âš™ï¸ Adding simulated column: {col_name}")
        if source_col in tmp.columns:
            tmp = tmp.withColumn(col_name, (col(source_col) + (rand()*2 - 1)).cast(DoubleType()))
        else:
            tmp = tmp.withColumn(col_name, expr().cast(DoubleType()))

print("âœ… Sensor columns ensured (cache skipped for serverless).")
display(tmp.limit(5))

# Step 4ï¸âƒ£ : Clean and Convert Sensor Columns
sensor_cols = ["Sensor_Temperature_C", "Sensor_Soil_Moisture_%", "NDVI_Index", "PH_Value", "Irrigation_Level_%"]
for c in sensor_cols:
    tmp = tmp.withColumn(c, col(c).cast(DoubleType()))

tmp = tmp.na.drop(subset=sensor_cols)
print("âœ… Cleaned dataset and dropped incomplete sensor rows.")

# Step 5ï¸âƒ£ : Anomaly Detection using Z-Score
group_col = "Crop_Type" if "Crop_Type" in tmp.columns else "District"

stats = tmp.groupBy(group_col).agg(
    avg("Sensor_Soil_Moisture_%").alias("mean_soil"),
    stddev("Sensor_Soil_Moisture_%").alias("std_soil"),
    avg("Sensor_Temperature_C").alias("mean_temp"),
    stddev("Sensor_Temperature_C").alias("std_temp"),
    avg("NDVI_Index").alias("mean_ndvi"),
    stddev("NDVI_Index").alias("std_ndvi")
)

joined = tmp.join(stats, on=group_col, how="left")

joined = joined.withColumn("z_soil", (col("Sensor_Soil_Moisture_%") - col("mean_soil")) / (col("std_soil") + 0.001))
joined = joined.withColumn("z_temp", (col("Sensor_Temperature_C") - col("mean_temp")) / (col("std_temp") + 0.001))
joined = joined.withColumn("z_ndvi", (col("NDVI_Index") - col("mean_ndvi")) / (col("std_ndvi") + 0.001))

joined = joined.withColumn("Anomaly_Soil", (abs(col("z_soil")) > 2.5).cast(IntegerType()))
joined = joined.withColumn("Anomaly_Temp", (abs(col("z_temp")) > 2.5).cast(IntegerType()))
joined = joined.withColumn("Anomaly_NDVI", (abs(col("z_ndvi")) > 2.5).cast(IntegerType()))

anomalies = joined.filter((col("Anomaly_Soil") + col("Anomaly_Temp") + col("Anomaly_NDVI")) > 0)

# Save anomaly data
anomaly_table = "workspace.default.sensor_anomalies"
anomalies.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable(anomaly_table)
print(f"📂 Saved sensor anomalies table: {anomaly_table}")

display(anomalies.limit(10))

# Step 6ï¸âƒ£ : Label Data for Water Stress Prediction
moisture_threshold = 30.0
labeled = joined.withColumn("Water_Stress_Label", when(col("Sensor_Soil_Moisture_%") < moisture_threshold, 1).otherwise(0))

# Step 7ï¸âƒ£ : Train Random Forest Classifier
feature_cols = ["Sensor_Soil_Moisture_%", "Sensor_Temperature_C", "NDVI_Index", "PH_Value", "Irrigation_Level_%"]
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
rf = RandomForestClassifier(featuresCol="features", labelCol="Water_Stress_Label", numTrees=50, maxDepth=5, seed=42)

pipeline = Pipeline(stages=[assembler, rf])
train_data, test_data = labeled.randomSplit([0.8, 0.2], seed=42)
rf_model = pipeline.fit(train_data)

# Evaluate Accuracy
preds = rf_model.transform(test_data)
evaluator = MulticlassClassificationEvaluator(labelCol="Water_Stress_Label", predictionCol="prediction", metricName="accuracy")
acc = evaluator.evaluate(preds)
print(f"âœ… Model trained successfully. Accuracy = {acc:.3f}")

# Step 8ï¸âƒ£ : Predict Water Stress for Entire Dataset
pred_full = rf_model.transform(labeled).withColumnRenamed("prediction", "Water_Stress_Prediction")

# Step 9ï¸âƒ£ : Aggregate & Generate Recommendations
agg_cols = ["District", "Crop_Type"] if "Farmer_ID" not in pred_full.columns else ["Farmer_ID", "District", "Crop_Type"]

summary = pred_full.groupBy(*agg_cols).agg(
    round(avg("Sensor_Soil_Moisture_%"), 2).alias("Avg_Soil_Moisture_%"),
    round(avg("Sensor_Temperature_C"), 2).alias("Avg_Temperature_C"),
    round(avg("NDVI_Index"), 2).alias("Avg_NDVI"),
    round(avg("PH_Value"), 2).alias("Avg_PH_Value"),
    round(avg("Irrigation_Level_%"), 2).alias("Avg_Irrigation_%"),
    sum(col("Anomaly_Soil") + col("Anomaly_Temp") + col("Anomaly_NDVI")).alias("Anomaly_Count"),
    count("*").alias("Total_Readings"),
    round(avg("Water_Stress_Prediction"), 2).alias("Water_Stress_Rate")
)

# Add Recommendations
summary = summary.withColumn(
    "Irrigation_Recommendation",
    when(col("Avg_Soil_Moisture_%") < 25, lit("IRRIGATE NOW"))
    .when(col("Avg_Soil_Moisture_%").between(25, 35), lit("MONITOR SOON"))
    .otherwise(lit("NO ACTION"))
)

summary = summary.withColumn(
    "Fertilizer_Recommendation",
    when(col("Avg_NDVI") < 0.4, lit("INCREASE FERTILIZER"))
    .when(col("Avg_NDVI").between(0.4, 0.6), lit("MAINTAIN LEVELS"))
    .otherwise(lit("REDUCE FERTILIZER"))
)

# Urgency Score (weighted)
summary = summary.withColumn(
    "Urgency_Score",
    round((lit(1) - col("Avg_Soil_Moisture_%")/100)*2 + col("Anomaly_Count")*0.5 + col("Water_Stress_Rate")*1.5, 2)
)

# Step 🔟 : Save Results to Delta Table
output_table = "workspace.default.precision_farming_results"
summary.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable(output_table)
print(f"📊 Precision farming summary saved to: {output_table}")

display(summary.limit(20))

# Step 1ï¸âƒ£1ï¸âƒ£ : Visualization - NDVI vs Soil Moisture
plot_pd = pred_full.select("NDVI_Index", "Sensor_Soil_Moisture_%").toPandas()
plt.figure(figsize=(7, 5))
plt.scatter(plot_pd["NDVI_Index"], plot_pd["Sensor_Soil_Moisture_%"], color="green", alpha=0.5)
plt.xlabel("NDVI Index")
plt.ylabel("Soil Moisture (%)")
plt.title("NDVI vs Soil Moisture (Smart Sensor Data)")
plt.grid(True)
display(plt)

# Step 1ï¸âƒ£2ï¸âƒ£ : Show Top Urgent Fields
top_urgent = summary.orderBy(col("Urgency_Score").desc()).limit(10)
display(top_urgent)

print("âœ… Precision Farming IoT & Smart Sensor Pipeline Completed Successfully")
print(f"Tables Created: {output_table}, {anomaly_table}")
```

```python
# =====================================================================
# 📘 Agri-Insurance Claim Fraud Detection (Databricks PySpark)
# Fully compatible with Serverless Compute
# =====================================================================

# Step 1ï¸âƒ£ : Import Libraries
from pyspark.sql import SparkSession
from pyspark.sql.functions import (
    col, when, lit, round, monotonically_increasing_id, count, sum as _sum,
    avg as _avg, stddev, to_date, concat, coalesce
)
from pyspark.sql.types import DoubleType, IntegerType, DateType
from pyspark.sql.window import Window
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import BinaryClassificationEvaluator
import matplotlib.pyplot as plt
import pandas as pd

# Initialize Spark
spark = SparkSession.builder.appName("AgriInsurance_Fraud_Detection").getOrCreate()

# Step 2ï¸âƒ£ : Load Dataset
input_table = "workspace.default.agriculture"
df = spark.table(input_table)
print(f"âœ… Loaded table: {input_table} | Rows: {df.count()}")
display(df.limit(10))

# Step 3ï¸âƒ£ : Cast Numeric Columns
for c in ["Yield_ton_per_acre", "Average_Price_INR", "Claim_Amount_INR", "Rainfall_mm"]:
    if c in df.columns:
        df = df.withColumn(c, col(c).cast(DoubleType()))

# Step 4ï¸âƒ£ : Handle Missing Claim_Amount_INR
from pyspark.sql.functions import rand
if "Claim_Amount_INR" not in df.columns:
    print("âš™ï¸ 'Claim_Amount_INR' missing â€” creating synthetic values.")
    df = df.withColumn(
        "Claim_Amount_INR",
        (coalesce(col("Yield_ton_per_acre"), lit(2.0)) *
         coalesce(col("Average_Price_INR"), lit(3000.0)) * (0.5 + rand())).cast(DoubleType())
    )

# Step 5ï¸âƒ£ : Create Synthetic Claim_Date (âœ… Fixed)
if "Claim_Date" in df.columns:
    df = df.withColumn("Claim_Date", to_date(col("Claim_Date")))
else:
    print("âš™ï¸ Creating synthetic 'Claim_Date' from 'Year'.")
    if "Year" in df.columns:
        df = df.withColumn("Claim_Date", to_date(concat(col("Year").cast("string"), lit("-06-30"))))
    else:
        df = df.withColumn("Claim_Date", to_date(lit("2023-06-30")))

# Step 6ï¸âƒ£ : Create Weather_Anomaly_Flag
if "Weather_Anomaly_Flag" not in df.columns:
    print("âš™ï¸ Creating 'Weather_Anomaly_Flag' using rainfall extremes.")
    if "Rainfall_mm" in df.columns:
        q = df.approxQuantile("Rainfall_mm", [0.05, 0.95], 0.01)
        low, high = q[0], q[1]
        df = df.withColumn(
            "Weather_Anomaly_Flag",
            when((col("Rainfall_mm") < low) | (col("Rainfall_mm") > high), lit(1)).otherwise(lit(0)).cast(IntegerType())
        )
    else:
        df = df.withColumn("Weather_Anomaly_Flag", lit(0).cast(IntegerType()))

# Step 7ï¸âƒ£ : Add Claim_ID if Missing
if "Claim_ID" not in df.columns:
    df = df.withColumn("Claim_ID", monotonically_increasing_id())

# Step 8ï¸âƒ£ : Feature Engineering
df = df.withColumn(
    "Claim_to_Yield_Ratio",
    (col("Claim_Amount_INR") / ((col("Yield_ton_per_acre") * col("Average_Price_INR")) + lit(1))).cast(DoubleType())
)

# âœ… Step 9ï¸âƒ£ : Compute Statistical Thresholds (Serverless Compatible)
mean_claim = float(df.agg(_avg("Claim_Amount_INR")).first()[0] or 0)
std_claim = float(df.agg(stddev("Claim_Amount_INR")).first()[0] or 0)
high_threshold = mean_claim + 2 * std_claim if std_claim and mean_claim else mean_claim * 1.5

df = df.withColumn(
    "High_Claim_Flag",
    when(col("Claim_Amount_INR") >= lit(high_threshold), lit(1)).otherwise(lit(0)).cast(IntegerType())
)

# Step 🔟 : Multiple Claims Flag
if "Farmer_ID" in df.columns:
    farmer_claims = df.groupBy("Farmer_ID").agg(count("*").alias("Farmer_Claim_Count"))
    df = df.join(farmer_claims, on="Farmer_ID", how="left")
    df = df.withColumn("Multiple_Claims_Flag", when(col("Farmer_Claim_Count") > 1, lit(1)).otherwise(lit(0)).cast(IntegerType()))
else:
    df = df.withColumn("Farmer_Claim_Count", lit(1)).withColumn("Multiple_Claims_Flag", lit(0))

# Step 1ï¸âƒ£1ï¸âƒ£ : Create Rule-Based Fraud Label
if "Fraud_Flag" not in df.columns:
    print("âš™ï¸ Creating 'Fraud_Flag' using rule-based logic.")
    df = df.withColumn(
        "Fraud_Flag",
        when(
            (col("High_Claim_Flag") == 1) & (col("Claim_to_Yield_Ratio") > 0.5), lit(1)
        ).when(
            (col("High_Claim_Flag") == 1) & (col("Weather_Anomaly_Flag") == 0) & (col("Claim_to_Yield_Ratio") > 0.2),
            lit(1)
        ).otherwise(lit(0)).cast(IntegerType())
    )

# Step 1ï¸âƒ£2ï¸âƒ£ : Feature Set for Model
feature_cols = [
    "Claim_to_Yield_Ratio",
    "High_Claim_Flag",
    "Weather_Anomaly_Flag",
    "Multiple_Claims_Flag",
    "Farmer_Claim_Count"
]
for opt in ["Yield_ton_per_acre", "Average_Price_INR", "Rainfall_mm"]:
    if opt in df.columns:
        feature_cols.append(opt)

model_df = df.select(feature_cols + ["Fraud_Flag", "Claim_ID", "Farmer_ID", "District", "Crop_Type", "Claim_Amount_INR", "Claim_Date"]).na.drop()
print(f"âœ… Model-ready data: {model_df.count()} rows, {len(feature_cols)} features")
display(model_df.limit(10))

# Step 1ï¸âƒ£3ï¸âƒ£ : Train Random Forest Model
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
rf = RandomForestClassifier(featuresCol="features", labelCol="Fraud_Flag", predictionCol="prediction", numTrees=100, maxDepth=6, seed=42)
pipeline = Pipeline(stages=[assembler, rf])

train, test = model_df.randomSplit([0.8, 0.2], seed=42)
model = pipeline.fit(train)

# Step 1ï¸âƒ£4ï¸âƒ£ : Evaluate Model
preds = model.transform(test)
evaluator = BinaryClassificationEvaluator(labelCol="Fraud_Flag", rawPredictionCol="rawPrediction", metricName="areaUnderROC")
auc = evaluator.evaluate(preds)
print(f"📊 Model AUC: {auc:.3f}")

# Step 1ï¸âƒ£5ï¸âƒ£ : Apply Model to Full Data
from pyspark.sql.functions import udf
def get_prob(v):
    try:
        return float(v[1])
    except:
        return 0.0
from pyspark.sql.types import DoubleType
get_prob_udf = udf(get_prob, DoubleType())

full_preds = model.transform(model_df)
full_preds = full_preds.withColumn("Fraud_Probability", get_prob_udf(col("probability"))) \
                       .withColumnRenamed("prediction", "Predicted_Fraud_Flag")

# Step 1ï¸âƒ£6ï¸âƒ£ : Save Predictions Table
pred_table = "workspace.default.insurance_fraud_predictions"
cols_to_save = ["Claim_ID","Farmer_ID","District","Crop_Type","Claim_Date","Claim_Amount_INR"] + feature_cols + ["Fraud_Probability","Predicted_Fraud_Flag","Fraud_Flag"]
full_preds.select(*[c for c in cols_to_save if c in full_preds.columns]) \
          .write.mode("overwrite").option("overwriteSchema","true").saveAsTable(pred_table)
print(f"📂 Saved predictions to {pred_table}")
display(spark.table(pred_table).limit(10))

# Step 1ï¸âƒ£7ï¸âƒ£ : Save Suspected Fraud Cases (Probability â‰¥ 0.6)
suspected = full_preds.filter(col("Fraud_Probability") >= 0.6)
suspected_table = "workspace.default.suspected_fraud_cases"
suspected.select(*[c for c in cols_to_save if c in suspected.columns]) \
          .write.mode("overwrite").option("overwriteSchema","true").saveAsTable(suspected_table)
print(f"âš ï¸ Saved suspected fraud cases to {suspected_table}")
display(spark.table(suspected_table).limit(10))

# Step 1ï¸âƒ£8ï¸âƒ£ : Visualization
pdf = spark.table(pred_table).select("Fraud_Probability").toPandas()
plt.figure(figsize=(6,4))
plt.hist(pdf["Fraud_Probability"], bins=20, color="orange", edgecolor="black")
plt.title("Distribution of Fraud Probability")
plt.xlabel("Fraud Probability")
plt.ylabel("Claims Count")
display(plt)

print("âœ… Fraud detection pipeline executed successfully!")
```

```python
# ================================================================
# Farm-to-Fork Supply Chain Optimization - Final Fixed Notebook
# Input table: workspace.default.agriculture
# Output Delta tables:
#  - workspace.default.supply_chain_forecast
#  - workspace.default.inventory_recommendations
#  - workspace.default.supplier_scores
#  - workspace.default.route_clusters
#  - workspace.default.supply_chain_dashboard
# Serverless-compatible (no .cache(), no RDD.map stdev, concat used for dates)
# ================================================================

# Step 0: Imports
from pyspark.sql import SparkSession
from pyspark.sql.functions import (
    col, round, lit, avg as _avg, stddev as _stddev, sum as _sum,
    when, monotonically_increasing_id, to_date, concat, rand, year as _year,
    count
)
from pyspark.sql.types import DoubleType, IntegerType
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.clustering import KMeans
import matplotlib.pyplot as plt

# Initialize Spark
spark = SparkSession.builder.appName("Farm_to_Fork_SupplyChain_Optimization_Fixed").getOrCreate()

# Step 1: Load base dataset
input_table = "workspace.default.agriculture"
df = spark.table(input_table)
print(f"Loaded {input_table} rows: {df.count()}")
display(df.limit(5))
df.printSchema()

# Step 2: Ensure required supply-chain columns exist (simulate if missing)
# Columns we expect: Farm_ID, District, Crop_Type, Year, Production_tons, Avg_Price_INR,
# Supplier_ID, Supplier_Reliability (0-1), Lead_Time_days, Transport_Cost_per_ton, Demand_quintal, Latitude, Longitude

tmp = df

# Ensure Farm_ID exists
from pyspark.sql.functions import lit as _lit
if "Farm_ID" not in tmp.columns:
    tmp = tmp.withColumn("Farm_ID", (monotonically_increasing_id() % 10000).cast(IntegerType()))

# Production_tons - if we have Yield_ton_per_acre and Area_acres approximate total production
if "Production_tons" not in tmp.columns:
    if "Yield_ton_per_acre" in tmp.columns and "Area_acres" in tmp.columns:
        tmp = tmp.withColumn("Production_tons", (col("Yield_ton_per_acre") * col("Area_acres")).cast(DoubleType()))
    else:
        tmp = tmp.withColumn("Production_tons", ((10 + rand()*50)).cast(DoubleType()))

# Supplier_ID and Supplier_Reliability
if "Supplier_ID" not in tmp.columns:
    tmp = tmp.withColumn("Supplier_ID", (monotonically_increasing_id() % 20).cast(IntegerType()))
if "Supplier_Reliability" not in tmp.columns:
    tmp = tmp.withColumn("Supplier_Reliability", (0.6 + rand()*0.4).cast(DoubleType()))

# Lead time and transport cost
if "Lead_Time_days" not in tmp.columns:
    tmp = tmp.withColumn("Lead_Time_days", (3 + (rand()*14)).cast(IntegerType()))
if "Transport_Cost_per_ton" not in tmp.columns:
    tmp = tmp.withColumn("Transport_Cost_per_ton", (2000 + rand()*3000).cast(DoubleType()))

# Demand_quintal (market demand) - if not present, derive from Production_tons & random factor
if "Demand_quintal" not in tmp.columns:
    tmp = tmp.withColumn("Demand_quintal", (col("Production_tons") * 10 * (0.8 + rand()*0.6)).cast(DoubleType()))

# geo coordinates for simple routing
if "Latitude" not in tmp.columns or "Longitude" not in tmp.columns:
    tmp = tmp.withColumn("Latitude", (15 + rand()*5).cast(DoubleType()))
    tmp = tmp.withColumn("Longitude", (75 + rand()*5).cast(DoubleType()))

# Cast important fields
tmp = tmp.withColumn("Production_tons", col("Production_tons").cast(DoubleType()))
tmp = tmp.withColumn("Demand_quintal", col("Demand_quintal").cast(DoubleType()))
tmp = tmp.withColumn("Lead_Time_days", col("Lead_Time_days").cast(IntegerType()))
tmp = tmp.withColumn("Supplier_Reliability", col("Supplier_Reliability").cast(DoubleType()))

print("Simulated / ensured supply-chain columns.")
display(tmp.select(
    "Farm_ID","Production_tons","Supplier_ID","Supplier_Reliability","Lead_Time_days",
    "Transport_Cost_per_ton","Demand_quintal","Latitude","Longitude"
).limit(5))

# Step 3: Aggregate historical demand by Crop_Type and Year (for forecasting)
if "Year" not in tmp.columns:
    tmp = tmp.withColumn("Year", lit(2024))

agg = tmp.groupBy("Crop_Type", "Year").agg(
    _avg("Demand_quintal").alias("Avg_Demand_quintal"),
    _avg("Production_tons").alias("Avg_Production_tons")
).orderBy("Crop_Type","Year")

display(agg.limit(20))

# Step 4: Simple demand forecasting per crop using Linear Regression on Year
forecast_rows = []
crop_list = [r["Crop_Type"] for r in agg.select("Crop_Type").distinct().collect()]
for crop in crop_list:
    crop_df = agg.filter(col("Crop_Type")==crop).orderBy("Year")
    if crop_df.count() < 2:
        last = crop_df.orderBy(col("Year").desc()).limit(1).collect()
        if last:
            last_year = int(last[0]["Year"])
            last_demand = float(last[0]["Avg_Demand_quintal"])
            forecast_rows.append((crop, last_year+1, last_demand, "na_lr"))
        else:
            forecast_rows.append((crop, 2025, 100.0, "na_lr"))
        continue

    # Build small spark df for linear regression
    lr_train = crop_df.select("Year","Avg_Demand_quintal")
    assembler = VectorAssembler(inputCols=["Year"], outputCol="features")
    lr_df = assembler.transform(lr_train).select("features","Avg_Demand_quintal")
    lr = LinearRegression(featuresCol="features", labelCol="Avg_Demand_quintal", maxIter=50)
    lr_model = lr.fit(lr_df)
    next_year = int(crop_df.agg({"Year":"max"}).collect()[0][0]) + 1
    features_next = spark.createDataFrame([(next_year,)], ["Year"])
    features_next = assembler.transform(features_next).select("features")
    pred_next = lr_model.transform(features_next).collect()[0].prediction
    forecast_rows.append((crop, next_year, float(pred_next), "lr"))

# assemble forecast DataFrame
forecast_schema = ["Crop_Type","Forecast_Year","Forecast_Demand_quintal","Method"]
forecast_df = spark.createDataFrame(forecast_rows, schema=forecast_schema)
display(forecast_df)

# Save forecast table for dashboard
forecast_table = "workspace.default.supply_chain_forecast"
forecast_df.write.mode("overwrite").option("overwriteSchema","true").saveAsTable(forecast_table)
print(f"Saved forecast to {forecast_table}")

# Step 5: Compute Inventory metrics per farm
crop_prod = tmp.groupBy("Crop_Type").agg(_sum("Production_tons").alias("Total_Crop_Production_tons"))
farm_with_share = tmp.join(crop_prod, on="Crop_Type", how="left") \
    .withColumn("Farm_Production_Share", when(col("Total_Crop_Production_tons") > 0, col("Production_tons") / col("Total_Crop_Production_tons")).otherwise(lit(0.0)))

# join forecast demand to farm
fc = forecast_df.select(col("Crop_Type").alias("FC_Crop"), col("Forecast_Year"), col("Forecast_Demand_quintal"))
farm_fc = farm_with_share.join(fc, farm_with_share.Crop_Type == fc.FC_Crop, how="left") \
    .drop("FC_Crop")

farm_fc = farm_fc.withColumn("Forecast_Year", when(col("Forecast_Year").isNull(), lit(2025)).otherwise(col("Forecast_Year")))
farm_fc = farm_fc.withColumn("Farm_Forecast_Demand_quintal", (col("Farm_Production_Share") * col("Forecast_Demand_quintal")).cast(DoubleType()))
farm_fc = farm_fc.withColumn("Farm_Forecast_Demand_tons", (col("Farm_Forecast_Demand_quintal")/10.0).cast(DoubleType()))

# Inventory calculations
z = 1.65  # service level z-score
farm_fc = farm_fc.withColumn("Annual_Demand_tons", (col("Farm_Forecast_Demand_tons").cast(DoubleType())))
farm_fc = farm_fc.withColumn("Daily_Demand_tons", when(col("Annual_Demand_tons")>0, col("Annual_Demand_tons")/lit(365.0)).otherwise(lit(0.001)))
# sigma approximation: 20% of daily demand scaled by sqrt(lead time)
farm_fc = farm_fc.withColumn("Sigma_LT", ( (col("Lead_Time_days")**0.5) * (0.2 * col("Daily_Demand_tons")) ))
farm_fc = farm_fc.withColumn("Safety_Stock_tons", (lit(z) * col("Sigma_LT")).cast(DoubleType()))
farm_fc = farm_fc.withColumn("ROP_tons", (col("Daily_Demand_tons") * col("Lead_Time_days") + col("Safety_Stock_tons")).cast(DoubleType()))

# EOQ simplified
unit_value = 20000.0
farm_fc = farm_fc.withColumn("Ordering_Cost_S", (col("Transport_Cost_per_ton") * 0.1).cast(DoubleType()))
farm_fc = farm_fc.withColumn("Holding_Cost_H", (lit(0.10) * lit(unit_value)).cast(DoubleType()))

# avoid sqrt of null/zero by safe expression
farm_fc = farm_fc.withColumn("EOQ_tons", when((col("Annual_Demand_tons")>0) & (col("Ordering_Cost_S")>0) & (col("Holding_Cost_H")>0),
                                              ((2 * col("Annual_Demand_tons") * col("Ordering_Cost_S") / col("Holding_Cost_H"))**0.5).cast(DoubleType())
                                             ).otherwise(col("ROP_tons")))

farm_fc = farm_fc.withColumn("Recommended_Order_tons", when(col("EOQ_tons").isNull(), col("ROP_tons")).otherwise(
    when(col("EOQ_tons") > col("Annual_Demand_tons"), col("Annual_Demand_tons")).otherwise(
        when(col("EOQ_tons") < col("ROP_tons"), col("ROP_tons")).otherwise(col("EOQ_tons"))
    )
))

farm_fc = farm_fc.withColumn("Safety_Stock_Ratio", when(col("Annual_Demand_tons")>0, col("Safety_Stock_tons") / col("Annual_Demand_tons")).otherwise(lit(0.0)))
farm_fc = farm_fc.withColumn("Stockout_Risk_Score", ( (1 - col("Supplier_Reliability"))*0.6 + (1 - col("Safety_Stock_Ratio"))*0.4 ).cast(DoubleType()))

# Round numeric columns for presentation
to_round = ["Production_tons","Farm_Forecast_Demand_tons","Annual_Demand_tons","Daily_Demand_tons","Safety_Stock_tons","ROP_tons","EOQ_tons","Recommended_Order_tons","Stockout_Risk_Score"]
for c in to_round:
    farm_fc = farm_fc.withColumn(c, round(col(c),2))

# Persist inventory recommendations table
inv_table = "workspace.default.inventory_recommendations"
inv_cols = ["Farm_ID","District","Crop_Type","Annual_Demand_tons","Daily_Demand_tons","Safety_Stock_tons","ROP_tons","EOQ_tons","Recommended_Order_tons","Stockout_Risk_Score","Supplier_ID","Supplier_Reliability","Lead_Time_days","Transport_Cost_per_ton"]
available_cols = [c for c in inv_cols if c in farm_fc.columns]
farm_fc.select(*available_cols).write.mode("overwrite").option("overwriteSchema","true").saveAsTable(inv_table)
print(f"Saved inventory recommendations to {inv_table}")
display(spark.table(inv_table).limit(10))

# Step 6: Supplier scoring
supplier_scores = farm_fc.groupBy("Supplier_ID").agg(
    _avg("Supplier_Reliability").alias("Avg_Supplier_Reliability"),
    _avg("Lead_Time_days").alias("Avg_Lead_Time_days"),
    _avg("Transport_Cost_per_ton").alias("Avg_Transport_Cost_per_ton"),
    _avg("Stockout_Risk_Score").alias("Avg_Supplier_Risk")
)

# Normalization values
range_stats = supplier_scores.agg(
    _sum(col("Avg_Supplier_Reliability")*0).alias("dummy")  # placeholder to ensure non-empty query; will be replaced properly
).collect()  # ensure table exists

# compute min/max safely
from pyspark.sql.functions import min as _min, max as _max
rs = supplier_scores.agg(
    _min("Avg_Supplier_Reliability").alias("min_rel"),
    _max("Avg_Supplier_Reliability").alias("max_rel"),
    _min("Avg_Lead_Time_days").alias("min_lt"),
    _max("Avg_Lead_Time_days").alias("max_lt"),
    _min("Avg_Transport_Cost_per_ton").alias("min_tc"),
    _max("Avg_Transport_Cost_per_ton").alias("max_tc")
).collect()[0]

min_rel, max_rel = rs["min_rel"], rs["max_rel"]
min_lt, max_lt = rs["min_lt"], rs["max_lt"]
min_tc, max_tc = rs["min_tc"], rs["max_tc"]

from pyspark.sql.functions import expr
def norm_expr_s(colname, mn, mx):
    if mn is None or mx is None or mn == mx:
        return lit(0.5)
    else:
        return (col(colname) - lit(mn)) / (lit(mx) - lit(mn))

supplier_scores = supplier_scores.withColumn("Rel_norm", norm_expr_s("Avg_Supplier_Reliability", min_rel, max_rel)) \
    .withColumn("LT_norm", norm_expr_s("Avg_Lead_Time_days", min_lt, max_lt)) \
    .withColumn("TC_norm", norm_expr_s("Avg_Transport_Cost_per_ton", min_tc, max_tc))

supplier_scores = supplier_scores.withColumn("Supplier_Score", round(0.6*col("Rel_norm") + 0.2*(1-col("LT_norm")) + 0.2*(1-col("TC_norm")),3))

supplier_table = "workspace.default.supplier_scores"
supplier_scores.select("Supplier_ID","Avg_Supplier_Reliability","Avg_Lead_Time_days","Avg_Transport_Cost_per_ton","Supplier_Score") \
    .write.mode("overwrite").option("overwriteSchema","true").saveAsTable(supplier_table)
print(f"Saved supplier scores to {supplier_table}")
display(spark.table(supplier_table).limit(20))

# Step 7: Route clustering - cluster farms by geo coords
num_distinct = tmp.select("District").distinct().count()
k = min(10, max(2, num_distinct))
coords_df = tmp.select("Farm_ID","Latitude","Longitude").na.drop().withColumn("lat", col("Latitude")).withColumn("lon", col("Longitude"))
assembler = VectorAssembler(inputCols=["lat","lon"], outputCol="features")
coords_feat = assembler.transform(coords_df).select("Farm_ID","features")

kmeans = KMeans(k=k, seed=42)
kmodel = kmeans.fit(coords_feat)
clusters = kmodel.transform(coords_feat).select("Farm_ID","prediction")

route_table = "workspace.default.route_clusters"
clusters.write.mode("overwrite").option("overwriteSchema","true").saveAsTable(route_table)
print(f"Saved route clusters to {route_table}")
display(spark.table(route_table).limit(20))

# Step 8: Combine recommendations into a single dashboard table
dashboard_df = farm_fc.select(
    "Farm_ID","District","Crop_Type","Production_tons","Supplier_ID","Supplier_Reliability",
    "Annual_Demand_tons","Daily_Demand_tons","Safety_Stock_tons","ROP_tons","Recommended_Order_tons","Stockout_Risk_Score"
).join(clusters, on="Farm_ID", how="left")

dash_table = "workspace.default.supply_chain_dashboard"
dashboard_df.write.mode("overwrite").option("overwriteSchema","true").saveAsTable(dash_table)
print(f"Saved combined dashboard table to {dash_table}")
display(spark.table(dash_table).limit(20))

# Step 9: Quick visual checks (Pandas)
fc_pd = forecast_df.toPandas()
plt.figure(figsize=(8,4))
plt.bar(fc_pd["Crop_Type"], fc_pd["Forecast_Demand_quintal"])
plt.xticks(rotation=45)
plt.title("Forecast demand (next year) by crop (quintal)")
plt.ylabel("Forecast Demand (quintal)")
plt.grid(axis='y')
display(plt)

inv_pd = spark.table(inv_table).orderBy(col("Stockout_Risk_Score").desc()).limit(10).toPandas()
plt.figure(figsize=(8,3))
plt.bar(inv_pd.get("Farm_ID").astype(str), inv_pd.get("Stockout_Risk_Score"))
plt.title("Top farms by stockout risk")
plt.xlabel("Farm_ID")
plt.ylabel("Stockout Risk Score")
display(plt)

sup_pd = spark.table(supplier_table).toPandas()
plt.figure(figsize=(6,3))
plt.hist(sup_pd["Supplier_Score"].fillna(0), bins=10)
plt.title("Supplier Score Distribution")
plt.grid(axis='y')
display(plt)

print("âœ… Farm-to-Fork Supply Chain Optimization pipeline complete.")
print("Tables created:")
print(" -", forecast_table)
print(" -", inv_table)
print(" -", supplier_table)
print(" -", route_table)
print(" -", dash_table)
```
